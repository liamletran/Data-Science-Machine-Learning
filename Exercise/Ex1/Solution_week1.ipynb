```python
#Ex8

import pandas as pd
import numpy as np
import os
import sys
import random



def detail_input(input, method):

    if method == "last_letters":
        target_order = -2
        first_letters_word2 = obtain_letters(input, target_order) 
        return first_letters_word2    

    elif method == "first_letters":
        target_order = 0
        last_letters_word3 = obtain_letters(input, target_order)
        return last_letters_word3

    
def obtain_letters(input, target_letter_order):
    
    # Assign the order of the second target letter in the current string
    second_target_letter_order = target_letter_order + 1
    
    # Create a dictionary from input where each character is mapped to its index
    detail_dict  = {i: j for i, j in enumerate(input)}
    
    # Target the last two letters in the current string
    last_letters = detail_dict[list(detail_dict.keys())[target_letter_order]] + detail_dict[list(detail_dict.keys())[second_target_letter_order]]
    
    return last_letters


        
def look_up_word(df_content, target_string, method = ''):
            
    # Define if we looking up for the words that starts or ends with a certain string
    if method == 'first_letters':
        # Look up the suitalbe words that start with the determined string 
        df_second_word = df_content[df_content[0].str.startswith(target_string)]
        return df_second_word

    elif method == 'last_letters':
        # Look up the suitalbe words that end with the determined string 
        df_third_word = df_content[df_content[0].str.endswith(target_string)]
        return df_third_word

            
            
############ MAIN #################        
            
            
# Change directory
path = 'C:/Users/Liam/Desktop/DS+ML/Exercise/Ex1'
os.chdir(path)


def main():
    # Define the two user's input words
    input1 = input('Please enter the start word: ')
    input4 = input('Please enter the goal word: ')
    
    # Assign empty strings for the second and third word
    input2, input3 = '', ''
    
    # Map the input words into the desired order
    list_word = [input1, input2, input3, input4]
    
    # Input file and 2 input words form user 
    filename = 'wordsEN.txt'

    try:
        # Open the file and read its content.
        with open(filename) as f:
            # Read the file and split the lines at the the enter 
            content = f.read().splitlines()
            
            # Put the content in a dataframe for fast data processing 
            df_content = pd.DataFrame(content)
    except FileNotFoundError:
        print(f'Error: File {filename} not found.')
   
    # Call the function to obtain the last letter of the first input word
    first_letters_word2 = detail_input(input1, method = 'last_letters')
    
    # Look up the third word
    last_letters_word3 = detail_input(input4, method = 'first_letters')

    
    # Look up the second word that contains the English words in the text file "wordsEn.txt"
    df_word2 = look_up_word(df_content, first_letters_word2, method = 'first_letters')
    df_word3 = look_up_word(df_content, last_letters_word3, method = 'last_letters')
    
    # Extract the last two letters of each word in df1
    df_word2['last_two'] = df_word2[0].str[-2:]

    # Extract the first two letters of each word in df2
    df_word3['first_two'] = df_word3[0].str[:2]

    # Merge the two dataframes based on the condition that last_two letters of df1 match first_two letters of df2
    merged_df = pd.merge(df_word2, df_word3, left_on='last_two', right_on='first_two', how='inner')
    
    # Filter the merged dataframe to include only word pairs with lengths greater than 3
    merged_df = merged_df[(merged_df['0_x'].str.len() > 2) & (merged_df['0_y'].str.len() > 2)]
 
    print(f"The shortest chain is: \n")
    print(input1)
    
    # If the filtered dataframe is not empty
    if not merged_df.empty:
        # Select a random pair from the filtered dataframe
        random_pair = merged_df.sample(1)

        for index, row in random_pair.iterrows():
            x_value = row['0_x']
            y_value = row['0_y']
            print(x_value)
            print(y_value)
    
    print(input4)
if __name__ == "__main__":
    main()
```
